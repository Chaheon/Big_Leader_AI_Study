```python
df.set_index('PassengerId', inplace=True)
df
```

```python
unique_dic = {}
for i in df.columns :
    unique_dic[i] = len(list(df[i].unique()))

unique_dic
```

```python
del_col = []
for i in unique_dic.keys() :
    if unique_dic[i]> 200 :
        df.drop(i, axis=1, inplace=True)
        del_col.append(i)

print(del_col)
df
```
